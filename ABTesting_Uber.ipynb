```python
from scipy import stats

# Define a function to perform t-tests and return results
def perform_t_tests(df1, df2, variable):
    return stats.ttest_ind(df1[variable], df2[variable], equal_var=False)
```

```python
t_tests_results = {}
variables_to_test = ['total_double_matches',
    'trips_pool', 'trips_express', 'rider_cancellations',
    'cost_per_trip', 'match_rate', 'double_match_rate', 'total_trips', 'total_matches'
]

for variable in variables_to_test:
    t_tests_results[variable] = perform_t_tests( df_5min, df_2min, variable) #5min treatment, 2min control


t_tests_results
```

```python
from statsmodels.formula.api import ols

# Define the list of outcome variables to analyze
outcome_variables = [
    'total_trips','trips_pool', 'trips_express', 'rider_cancellations',
    'cost_per_trip', 'match_rate', 'double_match_rate'
]

# Perform simple linear regression for each outcome variable
regression_results = {}

for variable in outcome_variables:
    # Perform the regression using OLS
    model = ols(f'{variable} ~ treat', data=df).fit()

    # Extract the coefficient for 'treat', its p-value, and the R-squared value of the model
    coefficient = model.params['treat']
    p_value = model.pvalues['treat']
    # r_squared = model.rsquared

    # Store these values in the dictionary
    regression_results[variable] = {'Coefficient': coefficient, 'p-value': p_value}

# Convert the results dictionary to a DataFrame for better presentation
regression_summary_df = pd.DataFrame(regression_results).T

regression_summary_df
```

```python
import statsmodels.api as sm
def multi_regression(df, y, x):
  X = df[x]
  X = sm.add_constant(X)
  Y = df[f'{y}']
  # Fit the regression model
  model = sm.OLS(Y, X).fit()
  coefficients = model.params
  p_values = model.pvalues
  return model.summary()
```
